# Visualizing the Dataframe with Pandas in the Jupyter notebook

```python
import sqlite3
import pandas as pd
```

## Establishing a connection to the database

```python
conn = sqlite3.connect('employee.db')
```

## Loading the employee database into pandas dataframe

```python
df = pd.read_sql_query("SELECT * FROM employee", con=conn, index_col='ID')
```

## Content of the Dataframe

```python
df
```

## Filtering on surname with pandas

```python
df[df['Surname']=='Ogundepo']
```

```python
df[df['Surname'].isin(['Ogundepo'])]
```

```python
df.loc[df['Surname']=='Olaseun', :]
```

```python
df.loc[:, "Middle_name":"Surname"]
```

### Sorting the index (ascending)

```python
df.sort_index()
```

## Closing the database

```python
conn.close()
```
